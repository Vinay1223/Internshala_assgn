```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine relevant text features for similarity (title + category name)
video_df['combined_features'] = video_df['title'] + ' ' + video_df['category.name']

# Create the TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(video_df['combined_features'])

# Compute cosine similarity between all videos
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Display the cosine similarity matrix
cosine_sim[:5, :5]
```

```python
def get_recommendations(user, video_df, cosine_sim):
    # Find the indices of the videos the user has watched
    watched_video_indices = [video_df[video_df['id'] == vid].index[0] for vid in user['watch_history']]
    
    # Calculate the mean similarity score of all videos against the watched videos
    similarity_scores = cosine_sim[watched_video_indices].mean(axis=0)
    
    # Sort videos by similarity score
    similar_videos_indices = similarity_scores.argsort()[::-1]
    
    # Recommend the top 5 most similar videos that the user hasn't watched yet
    recommendations = video_df.iloc[similar_videos_indices].head(10)
    recommendations = recommendations[~recommendations['id'].isin(user['watch_history'])]
    
    return recommendations[['id', 'title', 'category.name', 'view_count', 'average_rating']]

# Example: Get recommendations for User 1
recommendations_for_user_1 = get_recommendations(user_df.iloc[0], video_df, cosine_sim)
recommendations_for_user_1
```

```python
interaction_matrix = pd.DataFrame(0, index=user_df['user_id'], columns=video_df['id'])

for index, user in user_df.iterrows():
    interaction_matrix.loc[user['user_id'], user['watch_history']] = 1

interaction_matrix.head()
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(interaction_matrix)


distances, indices = model_knn.kneighbors(interaction_matrix, n_neighbors=2)


def collaborative_filtering_recommendation(user_id, interaction_matrix, video_df, n_recommendations=5):
    similar_users = indices[user_id]
    similar_users_videos = interaction_matrix.iloc[similar_users].sum(axis=0)
    recommended_videos = similar_users_videos.sort_values(ascending=False).head(n_recommendations)
    return video_df[video_df['id'].isin(recommended_videos.index)][['id', 'title', 'category.name']]


collaborative_recommendations = collaborative_filtering_recommendation(0, interaction_matrix, video_df)
collaborative_recommendations
```
